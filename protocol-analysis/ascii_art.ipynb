```python
DHBW_LOGO=( "              /| \r\n"
            "             /+| \r\n"
            " _________  /++|_____    _____  _    _ ____  _         __ \r\n"
            "|++++++++/||###|:::::|  |  __ `| |  | |  _ `| |  __   / / \r\n"
            "|+++++++/#||###|:::::|  | |  | | |__| | |_) | | /  | / / \r\n"
            "|++++++/##||##/::::::|  | |  | |  __  |  _ -| |/ | |/ / \r\n"
            "|+++++|###||#/:::::::|  | |__| | |  | | |_) | | /| | / \r\n"
            "|=====|###||/========|  |_____/|_|  |_|____/|__/ |__/ \r\n"
            "      |::/ \n"
            "      |:/ \n"
            "      |/ \n" )

TEST_STRING=( "underscore _ \n"
              "slash      / \n"
              "plus       + \n"
              "minus      - \n"
              "pipe       | \n"
              "open       ( \n"
              "close      ) \n"
              "hashtag    # \n"
              "double     : \n"
              "apastroph  ` \n")
```

```python
import serial, time

with serial.Serial(
    port='/dev/ttyUSB0',
    baudrate=9600,
    bytesize=serial.EIGHTBITS,    # Data Bits: 8
    stopbits=serial.STOPBITS_ONE, # Stop Bits: 1
    parity=serial.PARITY_NONE,    # Parity: None
    rtscts=True,                  # Hardware handshake
    timeout=1
) as con:

    # Normalize line endings
    # First, replace any existing \r\n with \n to avoid "double CRs" (\r\r\n)
    # Then, replace all \n with \r\n
    formatted_data = DHBW_LOGO.replace('\r\n', '\n').replace('\n', '\r\n')

    print("Sending data...")

    data = formatted_data.encode('ascii', errors='replace')

    data = data.replace(b'|', bytes([0x3C]))
    data = data.replace(b'`', bytes([0x60]))
    data = data.replace(b'#', bytes([0x7B]))

    # Send the data
    con.write(data)

    # Optional: Wait a moment to ensure transmission completes before closing
    time.sleep(1)
    print("Done.")
```
