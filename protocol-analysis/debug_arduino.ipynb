```python
from analysis_lib import KeysightInfiniiVisionMSO, DEFAULT_LABELS

ADDRESS = "USB0::2391::5925::MY49110266::INSTR"

DEBUG_LABELS = [
    "D_SI_PATH",
    "D_SO_PATH",
    "D_STATE",
    "D_KBRQ_F",
]

LABELS = DEFAULT_LABELS + DEBUG_LABELS


def setup_scope_debug(time_scale=1e-3, delay=2e-3, trigger_channel=3, trigger_slope="NEGative"):
    """
    Scope setup for handshake debugging
    """
    scope = KeysightInfiniiVisionMSO(ADDRESS)

    scope.setup_digital_channels(
        channels=range(len(LABELS)),
        threshold=2.5,
        time_scale=time_scale,
    )

    for i, label in enumerate(LABELS):
        scope.set_digital_label(i, label)

    scope.setup_digital_trigger(channel=trigger_channel, slope=trigger_slope)
    scope.set_trigger_delay(delay)

    return scope


# Trigger on READY falling edge, 1ms/div, 2ms delay
with setup_scope_debug(time_scale=1e-3, delay=2e-3, trigger_channel=3, trigger_slope="NEGative") as scope:
    pass
```
